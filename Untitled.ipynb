```python
import fastf1
import fastf1.plotting
import matplotlib.pyplot as plt
```

```python
# enable some matplotlib patches for plotting timedelta values and load
# FastF1's default color scheme
fastf1.plotting.setup_mpl()

# load a session and its telemetry data
session = fastf1.get_session(2021, 'Spanish Grand Prix', 'Q')
session.load()
```

```python
rbr_color = fastf1.plotting.team_color('RBR')
mer_color = fastf1.plotting.team_color('MER')

fig, ax = plt.subplots()
ax.plot(ver_tel['Distance'], ver_tel['Speed'], color=rbr_color, label='VER')
ax.plot(ham_tel['Distance'], ham_tel['Speed'], color=mer_color, label='HAM')

ax.set_xlabel('Distance in m')
ax.set_ylabel('Speed in km/h')

ax.legend()
plt.suptitle(f"Fastest Lap Comparison \n "
             f"{session.event['EventName']} {session.event.year} Qualifying")

plt.show()
```

```python
from scipy import signal
import numpy as np
import F1_Helper

speed_np = np.array(tel['Speed'])
differential = np.gradient(speed_np)
diff_x = np.arange(0, len(differential))

mins, maxes = F1_Helper.calc_min_max(speed_np, 0.7)

max_select = np.array(tel['Speed'].iloc[maxes])
min_select = np.array(tel['Speed'].iloc[mins])
max_dist = np.array(tel['Distance'].iloc[maxes])
min_dist = np.array(tel['Distance'].iloc[mins])
diff_x = np.array(tel['Distance'].iloc[diff_x])
diff2 = np.gradient(differential)
```

```python
rbr_color = fastf1.plotting.team_color('RBR')

fig, ax = plt.subplots(figsize=(20,20))
ax.plot(tel['Distance'], tel['Speed'], color=rbr_color, label='VER')
ax.scatter(max_dist, max_select, color='red', label='MAX', s=100)
ax.scatter(min_dist, min_select, color='blue', label='MIN', s=100)
ax.plot(diff_x, differential, color='green', label='Differential')
ax.plot(diff_x, diff2, color='white', label='Differential')
ax.set_xlabel('Distance in m')
ax.set_ylabel('Speed in km/h')

ax.legend()
plt.suptitle(f"Fastest Lap Comparison \n "
             f"{session.event['EventName']} {session.event.year} Qualifying")

plt.show()
```

```python
from scipy import stats

session = fastf1.get_session(2021, 2, 'R')
session.load()
laps = session.laps

max_speed_list = np.empty((1,0))
max_dist_list = np.empty((1,0))
min_speed_list = np.empty((1,0))
min_dist_list = np.empty((1,0))
max_count = []
min_count = []
for driver in session.drivers:
    laps = session.laps.pick_driver(driver)
    for idx, lap in laps.iterlaps():
        tel = lap.get_car_data().add_distance()
        speed_np = np.array(tel['Speed'])
        differential = np.gradient(speed_np)
        mins, maxes = F1_Helper.calc_min_max(speed_np, 0.7)
        max_select = np.array(tel['Speed'].iloc[maxes])
        min_select = np.array(tel['Speed'].iloc[mins])
        max_dist = np.array(tel['Distance'].iloc[maxes])
        min_dist = np.array(tel['Distance'].iloc[mins])
        max_speed_list = np.append(max_speed_list, max_select)
        max_dist_list = np.append(max_dist_list, max_dist)
        min_speed_list = np.append(min_speed_list, min_select)
        min_dist_list = np.append(min_dist_list, min_dist)
        max_count.append(len(maxes))
        min_count.append(len(mins))
        


average_len = np.average(max_count)
median_len = np.median(max_count)
mode_len, _ = stats.mode(max_count)
std_dev = np.std(max_count)

fig, ax = plt.subplots(figsize=(15,15))

ax.hist(max_count, bins=16, linewidth=0.5, edgecolor="white")
ax.set(xlim=(0, max(max_count)), xticks=np.arange(1, max(max_count), 10))

plt.show()

print('Statistic metrics for the distribution of local maxes is the followinq')
print(f'Average: {average_len}  Median: {median_len}  Mode: {mode_len[0]}  Standard Deviation: {std_dev}')


average_len = np.average(min_count)
median_len = np.median(min_count)
mode_len, _ = stats.mode(min_count)
std_dev = np.std(min_count)

fig, ax = plt.subplots(figsize=(15,15))

ax.hist(min_count, bins=16, linewidth=0.5, edgecolor="white")
ax.set(xlim=(0, max(min_count)), xticks=np.arange(1, max(min_count), 10))

plt.show()

print('Statistic metrics for the distribution of local maxes is the followinq')
print(f'Average: {average_len}  Median: {median_len}  Mode: {mode_len[0]}  Standard Deviation: {std_dev}')
```

```python
brake_gas_dist_avg = {}
for k, v in distance_dict.items():
    brake_gas_dist_avg[k] = {}
    brake_gas_dist_avg[k]['brake'] = np.average(v['max'])
    brake_gas_dist_avg[k]['gas'] = np.average(v['min'])

late_brakers = sorted(brake_gas_dist_avg.items(), key = lambda x: x[1]['brake'])
early_gas = sorted(brake_gas_dist_avg.items(), key = lambda x: x[1]['gas'])

print("The top 10 late brakers are: ")
for idx, item in enumerate(late_brakers):
    print(f"Driver  {session.results['FullName'][item[0]]}\t {item[1]['brake']}")
    if idx == 10:
        break
print("\nThe top 10 early on the gas are: ")
for idx, item in enumerate(early_gas):
    print(f"Driver {session.results['FullName'][item[0]]}\t {item[1]['gas']}")
    if idx == 10:
        break
```

```python
fig, ax = plt.subplots(figsize=(20,20))

def plot_drivers(driver, all_kmeans_dict):
    max_dist = all_kmeans_dict[driver]['max'].cluster_centers_.transpose()[0]
    max_select = all_kmeans_dict[driver]['max'].cluster_centers_.transpose()[1]
    min_dist = all_kmeans_dict[driver]['min'].cluster_centers_.transpose()[0]
    min_select = all_kmeans_dict[driver]['min'].cluster_centers_.transpose()[1]
    
    return max_dist, max_select, min_dist, min_select

driver = '33'
max_dist, max_select, min_dist, min_select = plot_drivers(driver, all_kmeans_dict)
ax.scatter(max_dist, max_select, color='red', label=driver, s=100)
driver = late_brakers[-1][0]
max_dist, max_select, min_dist, min_select = plot_drivers(driver, all_kmeans_dict)
ax.scatter(max_dist, max_select, color='blue', label=driver, s=100)
driver = 'average'
max_dist, max_select, min_dist, min_select = plot_drivers(driver, all_kmeans_dict)
ax.scatter(max_dist, max_select, color='white', label=driver, s=100)

    
ax.set_xlabel('Distance in m')
ax.set_ylabel('Speed in km/h')

ax.legend()
plt.suptitle(f"Fastest Lap Comparison \n "
             f"{session.event['EventName']} {session.event.year} Qualifying")

plt.show()
```
